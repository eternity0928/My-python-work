## map & filter

```python
items = [500, 30, 2000]
c = 30.8594
```

```python
s = []
for i in items:
    s .append(float(c)*i)
s
```

```python
egg = []
for item in items:
    p =item*c
    egg.append(p)
egg
```

```python
def exchange(p):
    return p*c
```

```python
exchange(500)
```

```python
map(exchange,items)
```

### map(函數,List)

```python
list(map(exchange,items)) #list是叫他列出來
```

```python
egg = list(map(exchange,items))
```

```python
egg
```

### filter

```python
grades = [50, 59, 65, 80, 100, 35, 40, 88]
```

```python
def spam(g):
    return g >= 60
```

```python
spam(50)
```

```python
spam(80)
```

```python
filter(spam, grades)
```

```python
list(filter(spam, grades))
```

### lambda

```python
list(filter(lambda g:g>=60, grades))
```
