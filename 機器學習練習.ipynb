```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0, 100, 40)
yy = np.linspace(0, 100, 40)
```

### 利用meshgrid產生格點

```python
X, Y = np.meshgrid(xx, yy)
```

```python
plt.scatter(X, Y)
```

### 所有的點 X,Y合併成P

```python
P = np.c_[X.ravel(), Y.ravel()]
```

```python
print(P)
```

## 模擬資料

```python
xr = np.random.randint(0, 100, 5)
yr = np.random.randint(0, 100, 5)
```

```python
plt.scatter(xr,yr)
```

```python
X_list = np.c_[xr,yr]
```

```python
X_list
```

```python
x = np.array(X_list)
```

```python
x
```

```python
y = np.array([1, 2, 3, 4, 5])
```

### 製作標準答案

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap = 'tab20b', s=100)
```

## 機器學習

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma = 'auto')
```

```python
clf.fit(x,y)
```

```python
y_predict = clf.predict(P)
```

```python
y_predict
```

```python
plt.scatter(P[:,0], P[:,1], c = y_predict, cmap = 'tab20b')
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap = 'tab20b', s=100)
plt.scatter(P[:,0], P[:,1], c = y_predict, cmap = 'tab20b', alpha = 0.3)
```

```python
x_arr = np.arange(-5, 105, 2.5)
y_arr = np.arange(-5, 105, 2.5)
X, Y = np.meshgrid(x_arr, y_arr)
P = np.c_[X.ravel(), Y.ravel()]
z = clf.predict(P)
Z = z.reshape(X.shape)

plt.contourf(X, Y, Z, cmap = 'tab20b', alpha = 0.3)
plt.scatter(x[:,0], x[:,1], c=y, cmap='tab20b', s=100)
```
