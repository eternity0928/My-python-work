### 變項名稱
    Attribute 1 : User
    Attribute 2 : art galleries
    Attribute 3 : Average user feedback on dance clubs
    Attribute 4 : Average user feedback on juice bars
    Attribute 5 : Average user feedback on restaurants
    Attribute 6 : Average user feedback on museums
    Attribute 7 : Average user feedback on resorts
    Attribute 8 : Average user feedback on parks/picnic spots
    Attribute 9 : Average user feedback on beaches
    Attribute 10 : Average user feedback on theaters
    Attribute 11 : Average user feedback on religious institutions

### 描述: 美國旅客至東亞旅遊對上述十個項目的評價
##### Reviews on destinations in 10 categories mentioned across East Asia.
### Excellent(4), Very Good(3), Average(2), Poor(1), and Terrible(0) and average rating is used.

```python
column_names = "C,art galleries,dance clubs,juice bars,restaurants,museums,resorts,parks/picnic spots,beaches,theaters,religious institutions".split(',')
```

```python
df= pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00484/tripadvisor_review.csv',names =  column_names, encoding = "UTF-8")
```

```python
# 刪掉不要的行
df= df.drop(0, axis = 0, inplace = False)
```

```python
df.head()
```

```python
#共有幾筆資料/資料數目長度
len(df)
```

```python
#資料框的基本資訊
df.info()
```

### 因為讀入的檔案資料內容為物件,先將數據轉為數值

```python
df['beaches'] = df['beaches'].astype('float')
```

```python
df.beaches.mean()
```

### 用for迴圈修改全部的數據

```python
for i in df[["art galleries","dance clubs","juice bars","restaurants","museums","resorts","parks/picnic spots","beaches","theaters","religious institutions"]]:
    df[i] = df[i].astype('float')
```

```python
df.head()
```

```python
df[:].beaches.mean()
```

```python
plt.subplot(3,4,1)
df['art galleries'].hist(bins=100,label = 'art galleries')
plt.subplot(3,4,2)
df['dance clubs'].hist(bins=100,label = 'dance clubs')
plt.subplot(3,4,3)
df['juice bars'].hist(bins=100,label = 'juice bars')
plt.subplot(3,4,4)
df['restaurants'].hist(bins=100,label = 'restaurants')
plt.subplot(3,4,5)
df['museums'].hist(bins=100,label = 'museums')
plt.subplot(3,4,6)
df['resorts'].hist(bins=100,label = 'resorts')
plt.subplot(3,4,7)
df['parks/picnic spots'].hist(bins=100,label = 'parks/picnic spots')
plt.subplot(3,4,8)
df['beaches'].hist(bins=100,label = 'beaches')
plt.subplot(3,4,9)
df['theaters'].hist(bins=100,label = 'theaters')
plt.subplot(3,4,10)
df['religious institutions'].hist(bins=100,label = 'religious institutions')
```

## 計算各個景點的平均評價

```python
smean = ["平均評價",]
for i in df[["art galleries","dance clubs","juice bars","restaurants","museums","resorts","parks/picnic spots","beaches","theaters","religious institutions"]]:
    smean.append(df[i].mean())
smean = np.array(smean)
```

```python
smean = smean.reshape(1,11)
```

### 製成DataFrame

```python
df2 = pd.DataFrame(smean, columns=column_names)
```

```python
df2
```

```python
df2.index = 'Summary'.split(',')
```

```python
df2
```

## DataFrame的合併

```python
df_c = pd.concat([df, df2], axis=0)
```

```python
df_c.tail()
```

```python
df_c.loc["Summary"]
```

### 由上述數據可以看出,美國旅客對東亞的公園/野餐景點的平均滿意度最高,對餐廳的滿意度最低
