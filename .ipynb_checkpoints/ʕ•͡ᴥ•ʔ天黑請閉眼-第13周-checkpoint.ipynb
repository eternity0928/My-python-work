```python
def g(n):
    print(n + "çš„èº«åˆ†æ˜¯" + predict[n])

def prophet():
    print("ä»Šæ™šä½ è¦æŸ¥é©—èº«åˆ†çš„æ˜¯")
    interact_manual(g, n = player_predict)

prophet()
```
